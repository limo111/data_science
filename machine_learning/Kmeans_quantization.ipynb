# K-means clustering can be used for color quantization, which involves reducing the number of colors used in an image while preserving the overall appearance of the image. This can be useful for reducing the file size of images or for simplifying the color palette for certain applications.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
img=mpimg.imread('../data/palm_trees.jpg')
```

```python
plt.imshow(img)
```

```python
(h,w,c)=img.shape
```

```python
image=img.reshape(h*w,c)
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=6)
```

```python
labels=model.fit_predict(image)
```

```python
rbf=model.cluster_centers_
```

```python
image=np.round(rbf[labels],0).astype(int)
```

```python
image=image.reshape(h,w,c)
```

```python
plt.imshow(image)
```
