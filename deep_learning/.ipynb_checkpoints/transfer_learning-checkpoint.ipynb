## Creating a CNN model

```python
model = Sequential()

model.add(Conv2D(filters=32, kernel_size=(3,3),input_shape=[130,130,3], activation='relu',))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=64, kernel_size=(3,3), activation='relu',))
model.add(MaxPooling2D(pool_size=(2, 2)))

model.add(Conv2D(filters=64, kernel_size=(3,3), activation='relu',))
model.add(MaxPooling2D(pool_size=(2, 2)))


model.add(Flatten())


model.add(Dense(128))
model.add(Activation('relu'))

# Dropouts help reduce overfitting by randomly turning neurons off during training.
# Here we say randomly turn off 50% of neurons.
model.add(Dropout(0.5))

# Last layer, remember its binary so we use sigmoid
model.add(Dense(1))
model.add(Activation('sigmoid'))

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
model.summary()
```

## Why use a pretrained model?
### Using a pretrained model is often preferred over training a model from scratch when you have limited labeled data or when the task at hand is similar to the pretraining task. Here are a few reasons why pretrained models are commonly preferred:

### Transfer learning: Pretrained models are typically trained on large-scale datasets, such as ImageNet, which contain millions of labeled images. By leveraging the knowledge learned from these datasets, pretrained models can provide a good starting point for similar tasks. The lower layers of a pretrained model learn generic features such as edges, textures, and basic shapes, which are beneficial for various computer vision tasks.

### Feature extraction: Pretrained models can serve as powerful feature extractors. Instead of training a model from scratch and learning features from raw pixel data, you can use the pretrained model to extract high-level features from your data. These features can then be fed into a smaller, custom-built model that is trained specifically for your task. This approach often leads to better performance with less training data.

### Faster convergence: Since pretrained models have already learned a set of meaningful features, they can significantly speed up the convergence of training. By starting with pretrained weights, the model is initialized closer to the optimal solution, allowing it to fine-tune its parameters more efficiently.

### Regularization: Pretrained models have been trained on large and diverse datasets, which helps them generalize well to new data. The knowledge acquired during pretraining acts as a form of regularization, reducing the risk of overfitting on small or limited training data.

## pretrained InceptionV3 model with weights pre-trained on the ImageNet dataset, excluding the top layer, and sets the input shape to [130, 130, 3].

```python
pretrained=InceptionV3(weights='imagenet',include_top=False,input_shape=[130,130,3])
```

## Freezes all layers in the pretrained model, making them untrainable.

```python
for layer in pretrained.layers:
    layer.trainable=False
```

```python
last_layer=pretrained.get_layer('mixed7')
```

```python
last_output=last_layer.output
```

## The above code retrieves the layer named 'mixed7' from the pretrained model and assigns it to the variable `last_layer`. The variable `last_output` stores the output of that layer, which represents the features extracted by the model up to that point.

```python
x=tensorflow.keras.layers.Flatten()(last_output)
x=tensorflow.keras.layers.Dense(128,activation='relu')(x)
x=tensorflow.keras.layers.Dense(1,activation='sigmoid')(x)
model=Model(pretrained.input,x)
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

## The above code defines additional layers on top of the extracted features from the `last_output`. It creates a new model that takes the pretrained model's input and passes it through a flatten layer, two dense layers, and outputs a single value. The model is then compiled with the Adam optimizer, binary cross-entropy loss, and accuracy metric.

```python
model.summary()
```

```python
early=EarlyStopping(monitor='val_loss',mode='min',patience=2)
```

```python
model.fit(train_gen,validation_data=[test_gen],epochs=20,callbacks=[early],steps_per_epoch=100,validation_steps=100)
```

## ## The code trains the new model using the training data generator (`train_gen`), validates it using the testing data generator (`test_gen`), for 20 epochs. It uses an early stopping callback and specifies the number of steps per epoch for training and validation.
